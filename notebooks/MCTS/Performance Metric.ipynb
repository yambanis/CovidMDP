```python
import pickle
import pandas as pd
from MCTS import mcts
import plotly.graph_objects as go
import numpy as np

with open("../../data/tests/mcts/sim_rewardexp_itL15H5d350.pkl", 'rb') as f:
    test_data = pickle.load(f)
```

```python
df = pd.DataFrame([pd.Series(d).value_counts() for d in test_data[0]]).fillna(0)
```

```python

fig = go.Figure()

hospital_beds = len(df[3])*[0.0025]
hospitalized = df[3]/55e3
x = df[3].index.to_list()

fig.add_trace(go.Scatter(x=x, y=hospitalized, name='hospitalized', line_color = 'purple'))
fig.add_trace(go.Scatter(x=x, y=hospital_beds, name='Unsued Capacity', line_color = 'orange', fill='tonexty'))
```

```python
score = 1 - (np.trapz(hospital_beds, x) - np.trapz(hospitalized, x))/ np.trapz(hospital_beds, x)
score
```
