```python
import networkx as nx
from networkx.drawing.nx_agraph import write_dot, graphviz_layout
import matplotlib.pyplot as plt
```

```python
policies = {'u': ['u', 'sd'], 
            'sd': ['u', 'sd', 'lq'],
            'lq': ['sd', 'lq', 'hq'],
            'hq': ['lq', 'hq', 'lck'],
            'lck': ['hq', 'lck']}
```

```python
level_1 = [('r', [x]) for x in policies.keys()]

level_2 = [(c,  c + [x])
                for p,c in level_1
                for x in policies[c[-1]]]

level_3 = [(c,  c + [x])
                for p,c in level_2
                for x in policies[c[-1]]]
```

```python
edge_list = level_1 + level_2 + level_3
edgelist = [("".join(x), "".join(y)) for x,y in edge_list]

G = nx.DiGraph(edgelist)
colors = [   'gray' if n.endswith('u')
             else 'green' if n.endswith('sd')
             else 'yellow' if n.endswith('lq')
             else 'orange' if n.endswith('hq')
             else 'red' if n.endswith('lck')
              else 'black' for n in G]
pos=graphviz_layout(G, prog='dot')
```

```python
for i in range(1, len(edgelist)):
    G = nx.DiGraph(edgelist[:i])
    c = colors[:i+1]
    p={k:v for k,v in list(pos.items())[:i+1]}
    _, ax = plt.subplots()
    print(p)
    nx.draw(G, pos=p, ax=ax, node_color=c, alpha=0.9, linewidths=1, edgecolors='black')
    plt.tight_layout();
    plt.savefig(f'action_plan_tree{i}.png', dpi=400, bbox_inches = 'tight',pad_inches = 0);    
```
